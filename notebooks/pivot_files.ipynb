```python
ROOT = Path('C:\\Users\\calvi\\Documents\\College\\UNCC\\Fall 2022\\ITSC 4155\\Project\\dataset\\transformed')
SOURCE_PATH = Path('C:\\Users\\calvi\\Documents\\College\\UNCC\\Fall 2022\\ITSC 4155\\Project\\dataset\\transformed\\counts_wireless_logs')
DEST_PATH = Path('C:\\Users\\calvi\\Documents\\College\\UNCC\\Fall 2022\\ITSC 4155\\Project\\dataset\\transformed\\counts_wireless_logs_pivot')
buildings = {'BURS','Biss','CenC','East','FacA','FouA','GRIG','McEn','NCRC','Nine','PROS','Scie'}

def convert_df(file_name):
    
    df = pd.read_csv(SOURCE_PATH / Path(file_name))
    df = df.pivot_table(values = 'count', index = 'timestamp', columns = 'building').reset_index()
    df = df.fillna(0)
    
    to_drop = buildings.intersection(set(df.columns))
    print(to_drop)
    if len(to_drop) > 0:
        df = df.drop(columns = to_drop)
    df.to_csv(DEST_PATH / Path(file_name))
    
```

```python
%%time
for f in files:
    convert_df(f)
```

### Checking that building names are consistent

```python
os.getcwd()
```

```python
os.chdir(DEST_PATH)
```

```python
files = os.listdir()
```

```python
column_set_list = []
for idx, f in enumerate(files):
    curr_df = pd.read_csv(f).drop(columns = 'Unnamed: 0')
    column_set_list.append(set(curr_df.columns))
    
    
```

```python
missing_buildings = set()
for col1 in column_set_list:
    for col2 in column_set_list:
        missing = col1.difference(col2)
        if len(missing) > 0:
            for m in missing:
                missing_buildings.add(m)
    
```

```python
missing_buildings
```

```python
df_list = []
for idx, f in enumerate(files):
    curr_df = pd.read_csv(f).drop(columns = 'Unnamed: 0')
    df_list.append(curr_df)
    
    
```

```python
# recursive function to concatenate dataframes :D 
def concat_df(df, idx):
    if idx == len(df_list) - 1:
        return df
    else:
        idx += 1
        df = pd.concat([df, concat_df(df_list[idx], idx)])
        return df
```

```python
cc_df = concat_df(df_list[0], 0)
```

```python
df = cc_df.sort_values(by = 'timestamp').reset_index().drop(columns=['index'])
```
