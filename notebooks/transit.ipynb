```python
def clean_df(transit):
    transit['datetime'] = transit['Date'] + ' ' + transit['Time']
    transit.datetime = pd.to_datetime(transit.datetime)
    transit = transit.drop(columns=['Date', 'Time'])
    transit = transit[transit['Route'].isin(['Green','Silver'])]
    return transit
```

```python
clean_dfs = []
for df in dfs:
    clean = clean_df(df)
    clean_dfs.append(clean)
```

```python
clean_dfs[3]
```

```python
# recursive function to concatenate dataframes :D 
def concat_df(df, idx):
    if idx == len(clean_dfs) - 1:
        return df
    else:
        idx += 1
        df = pd.concat([df, concat_df(clean_dfs[idx], idx)])
        return df
```

```python
full_df = concat_df(clean_dfs[0], 0)
```

```python
def transform_transit(df, route = None):
    if route is not None:
        df = df[df.Route == route]
    df = df[['On off', 'Count', 'datetime', 'Stop']]
    df['new_count'] = df['Count'] * -1
    df.loc[df['On off'] == 'on', 'new_count'] *= -1
    df = df.drop(columns = ['On off', 'Count'])
    df = df.pivot_table(values = 'new_count', index = 'datetime', columns = 'Stop').fillna(0).sort_index()
    df = df.groupby(df.index.strftime('%Y-%m-%d %H:00:00')).sum()
    return df
```
